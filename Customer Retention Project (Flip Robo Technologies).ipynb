### E-retail factors for customer activation and retention: A case study from Indian e-commerce customers

Customer satisfaction has emerged as one of the most important factors that guarantee the success of online store; it has been posited as a key stimulant of purchase, repurchase intentions and customer loyalty. A comprehensive review of the literature, theories and models have been carried out to propose the models for customer activation and customer retention. Five major factors that contributed to the success of an e-commerce store have been identified as: service quality, system quality, information quality, trust and net benefit. The research furthermore investigated the factors that influence the online customers repeat purchase intention. The combination of both utilitarian value and hedonistic values are needed to affect the repeat purchase intention (loyalty) positively. The data is collected from the Indian online shoppers. Results indicate the e-retail success factors, which are very much critical for customer satisfaction.

```python
# Let's import the necessary libraries for data analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Let's import the dataset for analysis

dataset = pd.read_excel("customer_retention_datasheet.xlsx")
dataset
```

```python
# Let's check the column names

dataset.columns = dataset.columns.str.strip()
dataset.columns
```

```python
# Let's check the value counts

for column in dataset:
    print(dataset[column].value_counts())
```

```python
# Let's check the Male and Female counts

sns.countplot(x='1Gender of respondent', data=dataset)
print(dataset["1Gender of respondent"].value_counts())
```

In this dataset most of the females are doing online shopping

```python
# Let's chec the age wise customer list

plt.figure(figsize=(15,5))
sns.countplot(x='2 How old are you?', data=dataset)
print(dataset['2 How old are you?'].value_counts())
```

Most of the customers aged between 21-50 years are doing online shopping

```python
# Let's check the most online shopping city

plt.figure(figsize=(15,5))
sns.countplot(x='3 Which city do you shop online from?', data=dataset)
print(dataset['3 Which city do you shop online from?'].value_counts())
```

Most of the online shopping customers are from NCR. So, need to focus on these cities.

```python
# Let's see who have been shopping online for how long ?

plt.figure(figsize=(15,4))
sns.countplot(x='5 Since How Long You are Shopping Online ?', data = dataset)
print(dataset['5 Since How Long You are Shopping Online ?'].value_counts())
```

Every age customer doing online shopping but Most of the customers are doing shopping more than 4 years. Let's check the age effect on it..

```python
# Let's see who have been shopping online for how long ?

plt.figure(figsize=(20,5))
sns.countplot(x='5 Since How Long You are Shopping Online ?', data = dataset, hue='2 How old are you?')
```

Now again it's clear that most of aged 21-50 years customers giving prefference to online shopping.

```python
# Online shopping trend in last 1 year

plt.figure(figsize=(15,4))
sns.countplot(x='6 How many times you have made an online purchase in the past 1 year?', data=dataset)
print(dataset['6 How many times you have made an online purchase in the past 1 year?'].value_counts())
```

It seems that shopping trend is very low in last year.

```python
# How do access the internet while shopping on-line?


plt.figure(figsize=(15,4))
sns.countplot(x='7 How do you access the internet while shopping on-line?', data=dataset)
print(dataset['7 How do you access the internet while shopping on-line?'].value_counts())
```

Most of the customers use their mobile internet for online shopping.

```python
# After first visit, how do customer reach the online retail store?                                            "

plt.figure(figsize=(15,4))
sns.countplot(x='13 After first visit, how do you reach the online retail store?',data=dataset)
print(dataset['13 After first visit, how do you reach the online retail store?'].value_counts())
```

All fields are looking good but Social Media is not playing good role in online shopping. We need to focus on this to increase the sale.

```python
# Customers prefered payment options

plt.figure(figsize=(10,4))
sns.countplot(x='15 What is your preferred payment Option?', data=dataset)
print(dataset['15 What is your preferred payment Option?'].value_counts())
```

Most of the customers preferred payment option as Credit/Debit Card, need to improve the CoD and E-wallet options to improve the sales.

```python
# Let's check the customer abandon

plt.figure(figsize=(20,8))
sns.countplot(x='17 Why did you abandon the “Bag”, “Shopping Cart”?', data=dataset)
print(dataset['17 Why did you abandon the “Bag”, “Shopping Cart”?'].value_counts())
```

```python
# Let's check the customer abandon with reason

plt.figure(figsize=(20,8))
sns.countplot(x='17 Why did you abandon the “Bag”, “Shopping Cart”?', data=dataset, hue='16 How frequently do you abandon (selecting an items and leaving without making payment) your shopping cart?')
print(dataset['17 Why did you abandon the “Bag”, “Shopping Cart”?'].value_counts())
```

Most of the customers leave the item with making the payment due to promocode not available and Better alternative offer available on other platforms

```python
# product Comparison

plt.figure(figsize=(10,4))
sns.countplot(x='19 Information on similar product to the one highlighted  is important for product comparison', data=dataset)
print(dataset['19 Information on similar product to the one highlighted  is important for product comparison'].value_counts())
```
